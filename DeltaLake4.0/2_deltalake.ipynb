# **Delta Log**

```python
#Making a data frame with hardcoded values and then displaying it
data = [(1, 100, 'aa'), (2, 200, 'bb'), (3, 300, 'cc'), (4, 400, 'dd')]
df = spark.createDataFrame(data, ['id', 'salary', 'name'])
display(df)
```

```python
#Code to write the data frame to a different directory
df.write.format("delta")\
    .mode("append")\
    .save("/Volumes/deltalaketest/default/deltavol/demosink/")
```

```python
#Code to read a specific file in a specific format and then display it
df=spark.read.format("json")\
    .load("/Volumes/deltalaketest/default/deltavol/demosink/_delta_log/00000000000000000001.json")
display(df)
```
