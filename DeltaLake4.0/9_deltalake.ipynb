# **Delta Lake Optimization**

```python
#create a data frame
df = spark.read.table("deltalaketest.default.clonetbl")
```

```python
display(df)
```

```python
#Code to overwrite the data frame.

df.write.format("delta")\
    .mode("append")\
    .save("/Volumes/deltalaketest/default/deltavol/optimization/")
```

```python
%sql
OPTIMIZE delta.`/Volumes/deltalaketest/default/deltavol/optimization/`
```

```python
%sql
DESCRIBE HISTORY delta.`/Volumes/deltalaketest/default/deltavol/optimization/`
```

```python
%sql
SELECT * FROM delta.`/Volumes/deltalaketest/default/deltavol/optimization/`
WHERE cust_id = 1
```

```python
%sql
OPTIMIZE delta.`/Volumes/deltalaketest/default/deltavol/optimization/` ZORDER BY (cust_id)
```
