# **Table Utility Commands**

### **Describe**

```python
%sql
DESCRIBE DETAIL deltalaketest.default.first_table

--Code to give details about the table
```

```python
%sql
DESCRIBE EXTENDED deltalaketest.default.first_table

--Code that gives more details about the table
```

### **Versioning**

```python
%sql
DESCRIBE HISTORY delta.`/Volumes/deltalaketest/default/deltavol/dmlsink/`

--Code to see the versioning of the table/folder.
```

### **Time Traveling**

```python
%sql
SELECT * FROM delta.`/Volumes/deltalaketest/default/deltavol/dmlsink/`
TIMESTAMP AS OF '2026-04-01T01:15:27.000+00:00'

--Code to time travel using the timestamp
```

```python
%sql
RESTORE delta.`/Volumes/deltalaketest/default/deltavol/dmlsink/`
TO VERSION AS OF 3

--Code to set the table back to version 3 via time traveling
```

### **Table Properties**

```python
%sql
SHOW TBLPROPERTIES deltalaketest.default.first_table

--Code to show the table properties
```

### **Vacuum**

```python
%sql
VACUUM delta.`/Volumes/deltalaketest/default/deltavol/dmlsink/` RETAIN 0 HOURS

--Code to delete parquet files. By default the data retention is 7 days. Always ask before using this command.
```

### **Cloning**

```python
%sql
CREATE TABLE deltalaketest.default.clonetbl
CLONE delta.`/Volumes/deltalaketest/default/deltavol/dmlsink/` VERSION AS OF 3

--Code to clone a table
```

```python
%sql
SELECT * FROM deltalaketest.default.clonetbl

--Query the cloned table
```

```python
%sql
CREATE TABLE deltalaketest.default.clonetblshallow
SHALLOW CLONE deltalaketest.default.first_table

--Code to shallow clone a table
```
