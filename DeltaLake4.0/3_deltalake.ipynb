# **Schema Enforcement**

```python
#Making a data frame with hardcoded values and then displaying it
data = [(1, 100, 'aa', 1), (2, 200, 'bb', 1), (3, 300, 'cc', 1), (4, 400, 'dd', 1)]
df = spark.createDataFrame(data, ['id', 'salary', 'name', 'sus_col'])
display(df)
```

```python
#Code to write the data frame to a different directory. This should fail in this case due to the schema mismatch
df.write.format("delta")\
    .mode("append")\
    .save("/Volumes/deltalaketest/default/deltavol/demosink/")
```

# **Schema Evolution**

```python
#Code to write the data frame to a different directory. Adding mergeSchema option to merge the schema and prevent the error above
df.write.format("delta")\
    .mode("append")\
    .option("mergeSchema", "true")\
    .save("/Volumes/deltalaketest/default/deltavol/demosink/")
```

# **Schema Overwrite**

```python
#Making a data frame with hardcoded values and then displaying it
data = [(1, 100, 'aa', 1), (2, 200, 'bb', 1), (3, 300, 'cc', 1), (4, 400, 'dd', 1)]
df = spark.createDataFrame(data, ['cust_id', 'income', 'name', 'tip'])
display(df)
```

```python
#Code to overwrite the data in the current location. You need to add overwriteSchema or it'll call a schema mismatch
df.write.format("delta")\
    .mode("overwrite")\
    .option("overwriteSchema", "true")\
    .save("/Volumes/deltalaketest/default/deltavol/demosink/")
```
