# **Delta Log Level Schema Changes**

```python
#Making an update to the data frame
data = [(1, 100, 'xyz', 1), (9, 200, 'bb', 1), (10, 300, 'cc', 1)]
df = spark.createDataFrame(data, ['cust_id', 'income', 'name', 'tip'])

display(df)
```

```python
#Code to save the data to a specific location
df.write.format("delta")\
    .mode("append")\
    .save("/Volumes/deltalaketest/default/deltavol/schemalevel/")
```

```python
%sql
-- Enable column mapping first. You need to do this otherwise the code below will error.
ALTER TABLE delta.`/Volumes/deltalaketest/default/deltavol/schemalevel/`
SET TBLPROPERTIES ('delta.columnMapping.mode' = 'name');

-- Code to rename a column in a table
ALTER TABLE delta.`/Volumes/deltalaketest/default/deltavol/schemalevel/`
RENAME COLUMN name TO customer_name
```
