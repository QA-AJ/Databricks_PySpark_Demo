# **Creating a Delta table**

### **Using SQL API**

```python
#imports
#Make sure you import fist. If you don't, sql functions and types will error out
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
%sql
CREATE TABLE deltalaketest.default.first_table
(
    id INT,
    salary INT
)
```

### **Using Delta API**

```python
#imports
#Make sure you import fist. If you don't, DeltaTable.createIfNotExists will error out
from delta.tables import DeltaTable, IdentityGenerator
```

```python
#Code to make a table using the Delta API
DeltaTable.createIfNotExists(spark) \
  .tableName("deltalaketest.default.firstdeltaapi") \
  .addColumn("id", "INT") \
  .addColumn("salary", "INT") \
  .execute()
```

```python
%sql
DROP TABLE deltalaketest.default.firstdeltaapi

--Deletes the table
```

# **Generated Columns**

### **Identity Columns**

```python
#Code to make a delta table using identity columns
DeltaTable.create(spark)\
  .tableName("deltalaketest.default.firstdeltaapi")\
  .addColumn("id_col", dataType=LongType(), generatedAlwaysAs=IdentityGenerator())\
  .addColumn("salary", dataType = IntegerType())\
  .addColumn("name", dataType = StringType())\
  .execute()
```

```python
%sql
DROP TABLE deltalaketest.default.firstdeltaapi

--Deletes the table
```

### **Computed Columns**

```python
#Code to make a delta table using computed columns
DeltaTable.create(spark)\
  .tableName("deltalaketest.default.firstdeltaapi")\
  .addColumn("salaryAfterTax", dataType=LongType(), generatedAlwaysAs="CAST((salary * 0.7) AS BIGINT)")\
  .addColumn("salary", dataType = IntegerType())\
  .addColumn("name", dataType = StringType())\
  .execute()
```
