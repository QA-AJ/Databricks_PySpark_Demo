# **DML**

```python
#Making a data frame with hardcoded values and then displaying it
data = [(5, 100, 'aa', 1), (6, 200, 'bb', 1), (7, 300, 'cc', 1), (8, 400, 'dd', 1)]
df = spark.createDataFrame(data, ['cust_id', 'income', 'name', 'tip'])

#Code to overwrite the data in the current location.
df.write.format("delta")\
    .mode("append")\
    .save("/Volumes/deltalaketest/default/deltavol/dmlsink/")
```

# **Upsert**

```python
#Making an update to the data frame
data = [(1, 100, 'xyz', 1), (9, 200, 'bb', 1), (10, 300, 'cc', 1)]
df = spark.createDataFrame(data, ['cust_id', 'income', 'name', 'tip'])

display(df)
```

```python
#Import
from delta.tables import DeltaTable
```

```python
#Code to declare the dlt_obj as the path
dlt_obj = DeltaTable.forPath(spark,"/Volumes/deltalaketest/default/deltavol/dmlsink")

#Code to merge the data to the new table via Upsert
dlt_obj.alias("trg").merge(df.alias("src"), "trg.cust_id = src.cust_id")\
    .whenMatchedUpdateAll()\
    .whenNotMatchedInsertAll()\
    .execute()
```
