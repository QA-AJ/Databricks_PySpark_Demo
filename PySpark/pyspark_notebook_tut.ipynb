# Spark Read CSV files

#### Basic read functions

```python
#Code to read CSV files with Spark.
df=spark.read.csv("/Volumes/my_files/my_schema/csv_files/my_files/sales.csv")

#Displays the data frame.
display(df)
```

```python
#Another way to read the CSV file.
df=spark.read.format("csv").load("/Volumes/my_files/my_schema/csv_files/my_files/sales.csv")

#Displays the data frame.
display(df)
```

```python
#Code to make the headers in the CSV file headers in the data frame.
df=spark.read.option("header", "true").csv("/Volumes/my_files/my_schema/csv_files/my_files/sales.csv")

#Displays the data frame.
display(df)
```

```python
#Code to make the data frame infer the correct data types in the data frame.
df=spark.read.option("header", "true").option("inferSchema", "true").csv("/Volumes/my_files/my_schema/csv_files/my_files/sales.csv")

#Displays the data frame.
display(df)
```

#### CSV file with no header

```python
#Code to provide the schema so we can have proper headers for the data frame while using a CSV file with no headers.
schema_str = "id INT, customer_name STRING, product STRING, quantity INT, price DOUBLE, remarks STRING"

#Accesses the schema code above to read the CSV file with no headers.
df=spark.read.schema(schema_str).csv("/Volumes/my_files/my_schema/csv_files/my_files/sales_no_header.csv")

#Displays the data frame.
display(df)
```

```python
#Second way to read a CSV file with no headers.
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, DoubleType

#Code for the schema to be used.
schema = StructType([
    StructField("id", StringType(), True),
    StructField("customer_name", StringType(), True),
    StructField("product", StringType(), True),
    StructField("quantity", IntegerType(), True),
    StructField("price", DoubleType(), True),
    StructField("remarks", StringType(), True)
])

#Access the schema code above to read the CSV file with no headers.
df=spark.read.schema(schema).csv("/Volumes/my_files/my_schema/csv_files/my_files/sales_no_header.csv")

#Displays the data frame.
display(df)
```

#### CSV file divided by |

```python
#Code to read a CSV file divied by pipe. Not doing this makes the entire data read as a single piece of data.
df=spark.read.option("header", "true").option("inferSchema", "true").option("delimiter", "|").csv("/Volumes/my_files/my_schema/csv_files/my_files/sales_pipe.csv")

#Displays the data frame.
display(df)
```

####CSV file with single ''

```python
#Code to read a CSV file with single '. You need to do this to remove the quote marks from the data otherwise it's not presented properly.
df=spark.read.option("header", "true").option("inferSchema", "true").option("delimiter", "|").option("quote", "'").csv("/Volumes/my_files/my_schema/csv_files/my_files/sales_pipe_single_quote.csv")

#Displays the data frame.
display(df)
```

#### CSV file with escape characters '#'

```python
#Code to read a CSV files with escape characters. You need to do this to remove the escape characters from the data otherwise it's not present properly.
df=spark.read.option("header", "true").option("inferSchema", "true").option("delimiter", "|").option("quote", "'").option("escape", "#").csv("/Volumes/my_files/my_schema/csv_files/my_files/sales_pipe_single_quote_escape.csv")

#Displays the data frame
display(df)
```

#### CSV file with multiple lines

```python
#Code to read a CSV file with multiple lines. You need to do this to show the correct data otherwise it's not present properly.
df=spark.read.option("header", "true").option("inferSchema", "true").option("delimiter", "|").option("quote", "'").option("escape", "#").option("multiLine", "true").csv("/Volumes/my_files/my_schema/csv_files/my_files/sales_pipe_single_quote_escape_multiline.csv")

#Displays the data frame
display(df)
```
