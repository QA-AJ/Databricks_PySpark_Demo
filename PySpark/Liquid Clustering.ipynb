```python
%sql
CREATE OR REPLACE TABLE my_files.my_schema.sales_partitioned (
order_id        STRING,
order_date      DATE,
country         STRING,
category        STRING,
amount          DOUBLE
)
USING DELTA
PARTITIONED BY (country, category)

--Creates a table
```

```python
%sql
INSERT INTO my_files.my_schema.sales_partitioned VALUES
('1', '2024-01-01', 'USA', 'Electronics', 1200.50),
('2', '2024-01-02', 'INDIA', 'Clothing', 450.00),
('3', '2024-01-03', 'USA', 'Clothing', 180.75),
('4', '2024-01-04', 'UK', 'Furniture', 899.00)

--Insert values into the created table
```

```python

df=spark.read.table("my_files.my_schema.sales_partitioned") #reads the table
df.write.partitionBy("country", "category").save("/Volumes/my_files/my_schema/test_volume/sales_partitioned") #partitions the table to a specific location
```

```python
%sql
CREATE OR REPLACE TABLE my_files.my_schema.sales_clustered (
order_id        STRING,
order_date      DATE,
country         STRING,
category        STRING,
amount          DOUBLE
)
USING DELTA
CLUSTER BY (country, category, order_date)

--Creates a table
```

```python
%sql
INSERT INTO my_files.my_schema.sales_clustered VALUES
('1', '2024-01-01', 'USA', 'Electronics', 1200.50),
('2', '2024-01-02', 'INDIA', 'Clothing', 450.00),
('3', '2024-01-03', 'USA', 'Clothing', 180.75),
('4', '2024-01-04', 'UK', 'Furniture', 899.00)

--Insert values into the created table
```
