# Transformations

```python
#Reads the table from the bakehouse samples
df=spark.read.table("samples.bakehouse.sales_franchises")

#displays the dataframe
display(df)
```

```python
%sql --converts the cell to SQL
SELECT * FROM samples.bakehouse.sales_franchises --displays the entire table
```

```python
#displays the columns
df.columns
```

```python
%sql
SHOW COLUMNS IN samples.bakehouse.sales_franchises --shows the columns
```

```python
#Select specific columns for the data frame.
df2=df.select("name", "country")

#displays the data.
display(df2)
```

```python
%sql
SELECT name, country FROM samples.bakehouse.sales_franchises --shows the name and country columns via SQL
```

```python
#import code
from pyspark.sql.functions import expr

#renames the column country as my_country
#df2=df.select("name",expr("country as my_country"))

#appends the city and country columns and lists them under the location column
df2=df.select("name",expr("concat(city,'_',country) as location"))

#displays the data.
display(df2)
```

```python
#import code
from pyspark.sql.functions import expr

#renames the column country as my_country
df2=df.selectExpr("name", "country as my_country")

#displays the data.
display(df2)
```

```python
#import code
from pyspark.sql.functions import expr

#appends the city and country columns and lists them under the location column using alias
df2=df.select("name",expr("concat(city,'_',country)").alias("location"))

#displays the data.
display(df2)
```

```python
%sql
SELECT name,concat(city,'_',country) AS location FROM samples.bakehouse.sales_franchises --appends the city and country columns and lists them under the location column using expr and alias
```

```python
#adds the location column to the table
df2=df.withColumn("location",expr("concat(city,'_',country)"))

#displays the data.
display(df2)
```

```python
%sql
SELECT *, concat(city,'_',country) AS location FROM samples.bakehouse.sales_franchises --appends the city and country columns and lists them under the location column using expr and alias
```

```python
#import code
from pyspark.sql.functions import lit

#Takes the data frame and adds a new column called one with a numeric value of 1.
df2=df.withColumn("one", lit(1))

#displays the data.
display(df2)
```

```python
%sql
SELECT *, 1 AS One FROM samples.bakehouse.sales_franchises --Takes the data frame and adds a new column called one with a numeric value of 1.
```

```python
#Code to drop the country column from the data frame.
df2=df.drop("country")

#displays the data.
display(df2)
```

```python

%sql
--Code to drop the country column from the data frame then display it.
SELECT * except(country) FROM samples.bakehouse.sales_franchises
```

```python
#Import code
from pyspark.sql.functions import col

#Select the column and changes the data type to double. You have to wrap the column in the col so the code knows it's a column.
df2=df.withColumn("supplier_id_double", col("supplierID").cast("double"))

#displays the data.
display(df2)
```

```python
%sql
SELECT *,cast(supplierID as double) AS supplier_id_double FROM samples.bakehouse.sales_franchises --Select the column and changes the data type to double. You have to wrap the column in cast so the code knows it's a column.
```
