```python
#code to read the data from a table and display it
df=spark.read.table("samples.bakehouse.sales_customers")
display(df)
```

```python
#Code to write the data to a volume
df.write.format("csv").save("/Volumes/my_files/my_schema/test_volume/test1")
```

```python
#writing the data as a table
df.write.saveAsTable("my_files.my_schema.sales_customers")
```

```python
#Overwrites the data that already exists
df.write.mode("overwrite").format("csv").save("/Volumes/my_files/my_schema/test_volume/test1")
```

```python
#Specifiying a partition
from pyspark.sql.functions import zip_with

df_with_address = df.withColumn("address", zip_with("phone_number"))
df_with_address.write.partitionBy("address").save("/Volumes/my_files/my_schema/test_volume/test2")
```
