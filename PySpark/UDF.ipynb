```python
# User Defined Function
data = [
    ("Order ID: 56789",),
    ("Invoice# 98321 processed",),
    ("No numbers here",),
    ("Ref-45A12B",)
]

df = spark.createDataFrame(data, ["text_col"])
df.createOrReplaceTempView("orders")
display(df)
```

```python
# COMMAND ----------

import re

# Define Python function for extraction
def extract_number(text):
    if text is None:
        return None
    match = re.search(r"\d+", text)   # regex to find first sequence of digits
    return match.group(0) if match else None

```

```python
# COMMAND ----------

spark.udf.register("extractNumberUDF", extract_number)
```

```python

%scala
import org.apache.spark.sql.SparkSession
import scala.util.matching.Regex

// Scala function to extract first numeric sequence from text
def extractNumber(text: String): String = {
   if (text == null) return null

  val pattern: Regex = "\\d+".r   // regex to find continuous digits
   pattern.findFirstIn(text) match {
     case Some(value) => value
     case None => null
   }
 }

// Register as UDF
spark.udf.register("extractNumberUDF", extractNumber _)
```

```python
from pyspark.sql.functions import col
from pyspark.sql.functions import expr

#Extracts the numbers from the data and places it into a different column. Anything with no number appears as null
df2=df.withColumn("extracted_number", expr("extractNumberUDF(text_col)"))

display(df2)
```

```python
%sql
 SELECT text_col,
          extractNumberUDF(text_col) AS extracted_number
   FROM orders
```
