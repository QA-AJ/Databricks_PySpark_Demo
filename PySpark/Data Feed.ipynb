```python
%sql
CREATE TABLE my_files.my_schema.people (
first_name       STRING,
age             LONG
)
USING DELTA TBLPROPERTIES (delta.enableChangeDataFeed = true)

--Creates a table and enables change data feed
```

```python

df=spark.createDataFrame([("Bob", 23), ("Sue", 25), ("Jim", 27)]).toDF(
    "first_name", "age"
) #creates a data frame based off the values above
df.write.mode("append").format("delta").saveAsTable("my_files.my_schema.people") #appends and saves the data
```

```python
%sql
SELECT * FROM my_files.my_schema.people

-- displays the data
```

```python
%sql
DELETE FROM my_files.my_schema.people WHERE first_name = 'Bob'

--deletes the row with the name bob
```

```python
%sql
UPDATE my_files.my_schema.people SET age=30 WHERE first_name = 'Jim'

--updates the row with the name jim
```

```python
#code to read the changes made from the previous cells. #0 for the start version shows everything
df=spark.read.format("delta")\
    .option("readChangeFeed", "true")\
    .option("startingVersion", 0)\ 
    .table("my_files.my_schema.people")
display(df)
```
